```python
import pandas as pd
```

```python
# gewünschte csv in session storage hochladen und name für pd.read_csv anpassen
data = pd.read_csv("datapull_stMargrethen.csv")
data.head()
```

```python
temperature_data = pd.json_normalize(data['Temperature'].apply(eval))
temperature_data.columns = ['temp min', 'temp max', 'temp afternoon', 'temp night', 'temp evening', 'temp morning']
data = pd.concat([data, temperature_data], axis=1)
data = data.drop(columns=['Temperature'])
data.head()
```

```python
humidity_data = pd.json_normalize(data['Humidity'].apply(eval))
humidity_data.columns = ['humidity']
data = pd.concat([data, humidity_data], axis=1)
data = data.drop(columns=['Humidity'])
data.head()
```

```python
rainfall_data = pd.json_normalize(data['Rainfall'].apply(eval))
rainfall_data.columns = ['rainfall']
data = pd.concat([data, rainfall_data], axis=1)
data = data.drop(columns=['Rainfall'])
data = data.drop(columns=['Wind Speed'])
data.head()
```

```python
data['temp average'] = data[['temp min', 'temp max', 'temp afternoon', 'temp night', 'temp evening']].mean(axis=1)
data.head()
```

```python
data['30-day rainfall cumulative'] = data['rainfall'].rolling(window=30).sum()
data['30-day temperature average'] = data['temp average'].rolling(window=30).mean()
data['30-day humidity average'] = data['humidity'].rolling(window=30).mean()
data.head(40)
```

```python
# nicht berücksichtigen für prediction
def calculate_critical_growth_state(row):
    temp_condition = 15 <= row['temp average'] <= 25
    rainfall_condition = 100 <= row['30-day rainfall cumulative'] <= 200
    humidity_condition = 50 <= row['humidity'] <= 70

    return temp_condition and rainfall_condition and humidity_condition

data['Critical Growth Stage'] = data.apply(calculate_critical_growth_state, axis=1)

data.head(20)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.sort_values(by=['Latitude', 'Longitude', 'Date'])

# saisonale Features
data['day_of_year'] = data['Date'].dt.dayofyear
data['month'] = data['Date'].dt.month
data['season'] = data['Date'].dt.month % 12 // 3 + 1  # 1=winter, 2=frühling, 3=sommer, 4=herbst

data.head()
```

```python
dataNew = data
def calculate_days_to_next_critical(df):
        """Calculate the days to the next 'Critical Growth Stage' for each row."""
        next_critical_dates = df.loc[df['Critical Growth Stage'], 'Date']
        df['days_to_next_critical'] = df['Date'].apply(
            lambda x: (next_critical_dates[next_critical_dates > x].min() - x).days
            if (next_critical_dates > x).any() else -1
        )
        return df

dataNew = dataNew.groupby(['Latitude', 'Longitude'], group_keys=False).apply(calculate_days_to_next_critical)
dataNew.head(40)
```

```python
dataNew.fillna(method='ffill', inplace=True)
dataNew.fillna(method='bfill', inplace=True)
dataNew.head(40)
```

```python
dataNew.drop(columns=['Latitude', 'Longitude', 'Date'])
dataNew.head()
```

```python
dataNew.to_csv("stMargrethenData_processed.csv", index=False)
```
