```python
import requests
import time
import pandas as pd
from datetime import datetime, timedelta

api_key = 'OpenWeatherMap_API_KEY'
endpoint = 'https://api.openweathermap.org/data/3.0/onecall/day_summary'

# gewünschte Koordinaten definieren
locations = [
    #{"lat": 47.3769, "lon": 8.5417} Zurich
    #{"lat": 46.8500, "lon": 9.5326} Chur
    # {"lat": 47.0486, "lon": 9.4419} Sargans
    # {"lat": 47.4245, "lon": 9.3767} St Gallen
    # {"lat": 46.1653, "lon": 8.7983} Locarno
    # {"lat": 46.9481, "lon": 7.4474} Bern
    # {"lat": 47.5596, "lon": 7.5886} # Basel
    # {"lat": 47.0502, "lon": 8.3093} # Luzern
    # {"lat": 47.1339, "lon": 9.0613} # Ziegelbrücke
    # {"lat": 47.2023, "lon": 8.7880} # Pfäffikon SZ
    {"lat": 47.4769, "lon": 9.6336} # St. Margrethen
]

start_date = datetime(2021, 1, 1)
end_date = datetime(2023, 12, 31)
dates = [start_date + timedelta(days=i) for i in range((end_date - start_date).days + 1)]

data = []

for location in locations:
    for date in dates:
        formatted_date = date.strftime("%Y-%m-%d")
        params = {
            'lat': location["lat"],
            'lon': location["lon"],
            'date': formatted_date,
            'appid': api_key,
            'units': "metric"
        }

        response = requests.get(endpoint, params=params)
        if response.status_code == 200:
            weather_data = response.json()
            data.append({
                "Latitude": location["lat"],
                "Longitude": location["lon"],
                "Date": formatted_date,
                "Temperature": weather_data.get("temperature"),
                "Humidity": weather_data.get("humidity"),
                "Wind Speed": weather_data.get("wind"),
                "Rainfall": weather_data.get("precipitation", 0)
            })
        else:
            print(f"Failed to retrieve data for {location} on {formatted_date}")

        time.sleep(1)  # API rate Limiten

df = pd.DataFrame(data)

# Daten speichern
df.to_csv("datapull_stMargrethen.csv", index=False)
print("Data saved")
```
